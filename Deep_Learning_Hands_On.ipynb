```python
# get the data

#upload the dataset to your Colab directory
df = pd.read_csv('data.csv')

df.head()
```

```python
#an information
df.info()
```

```python
# statiscal information about the data
df.describe()
```

```python
# drop unnecessary columns

df = df.drop(['id', 'Unnamed: 32'], axis = 1)

# df.drop(['id', 'Unnamed: 32'], axis = 1, inplace=True)
```

```python
#unneccessary columns have been dropped
df.head()
```

```python
df.diagnosis.value_counts()

#our dataset is fairly balanced
```

```python
sns.countplot(df.diagnosis)
```

```python
#encoding the target to numbers
to_replace = {'M': 1, 'B':0}

df.diagnosis.replace(to_replace, inplace=True)
```

```python
df.head()
```

```python
#seperating the features from the target
X = df.drop('diagnosis', axis=1)
y = df.diagnosis
```

```python
X.head()
```

```python
y
```

```python
#scaling the dataset using MinMax
#you can explore other options
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# from sklearn.model_selection import train_test_split

#data splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
# configuring the neural network

from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
input_dimension = X.shape[1]
input_dimension
```

```python
#assume this to be the forward propagation
model = Sequential([
    Dense(32, activation = 'relu', input_dim = input_dimension),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
  ])
```

```python
#assume this to be the backward prop.
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
#calulating the parameters
# (30 * 64) + 64
# (64 * 32) + 32
# (32 * 1) + 1
```

```python
# \training the model


history = model.fit(X_train, y_train, epochs = 100, validation_split=0.2)
```

```python
#overfitting
```

```python
tr_acc = history.history['accuracy']
tr_loss = history.history['loss']
val_acc = history.history['val_accuracy']
val_loss = history.history['val_loss']

epochs = [i+1 for i in range(len(tr_acc))]

plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs, tr_loss, 'r', label='Train Loss')
plt.plot(epochs, val_loss, 'g', label='Valid Loss')
plt.title('Loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(1, 2, 2)
plt.plot(epochs, tr_acc, 'r', label='Train Accuracy')
plt.plot(epochs, val_acc, 'g', label='Valid Accuracy')
plt.title('Accuracy')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
# \predicting the X_test
y_pred = model.predict(X_test)
```

```python
# made predictions 0 or 1
y_pred = (y_pred > 0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
# checked for accuracy
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# checking the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
